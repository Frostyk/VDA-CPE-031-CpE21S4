```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('heart_2020_cleaned.csv')

# Question 1: What is the distribution of age across different levels of heart disease?



plt.figure(figsize=(10, 6))
plt.hist(df[df['HeartDisease'] == 'Yes']['AgeCategory'], bins=20, alpha=0.5, label='Heart Disease')
plt.hist(df[df['HeartDisease'] == 'No']['AgeCategory'], bins=20, alpha=0.5, label='No Heart Disease')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age Across Heart Disease Status')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Observation: The visualization likely shows a higher frequency of heart disease among older age groups, indicating a correlation between age and the risk of heart disease.



```

```python
#Question 2: Does exercising regularly make a difference in how often people get heart disease?
plt.figure(figsize=(10, 6))
physical_activity_counts = df.groupby(['PhysicalActivity', 'HeartDisease'])['HeartDisease'].count().unstack()
physical_activity_counts.plot(kind='bar', stacked=True)
plt.xlabel('Physical Activity')
plt.ylabel('Count')
plt.title('Does exercising regularly make a difference in how often people get heart disease?')
plt.legend(title='Heart Disease')
plt.show()

# Observation: This visualization more clearly shows the percentage of people with heart disease for each physical activity category.
# It likely indicates that those who do not engage in physical activity have a higher chance of developing heart disease compared to those who do.
```

```python
#a. Read each device's date of releases and sales. Show it using a line plot
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Date of Release'], df['Price (USD)'], alpha=0.7, color='red')
plt.title('Prices over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#C
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

# Create the plot
plt.figure(figsize=(12, 6))  # Adjust figure size as needed

# Plot each product's sales
for product in df['Device'].unique():
    product_data = df[df['Device'] == product]
    plt.plot(product_data['Date of Release'], product_data['Price (USD)'], label=product)

# Customize the plot
plt.title('Sales of Technological Products Over Time')
plt.xlabel('Date of Release')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#D
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')


# Group the data by brand and sum the sales
brand_sales = df.groupby('Brand')['Price (USD)'].sum().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(brand_sales['Brand'], brand_sales['Price (USD)'])
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.title('Total Sales by Brand')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#E
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

# Group data by device and country of origin, then count occurrences
device_country_counts = df.groupby(['Device', 'Country of Origin'])['Device'].count().reset_index(name='Count')

# Get the most common country of origin for each device
most_common_countries = device_country_counts.loc[device_country_counts.groupby('Device')['Count'].idxmax()]

# Create the histogram
plt.figure(figsize=(12, 6))
plt.bar(most_common_countries['Device'], most_common_countries['Count'])
plt.xlabel('Device')
plt.ylabel('Frequency')
plt.title('Most Common Country of Origin for Each Device')
plt.xticks(rotation=45, ha='right')

# Add labels for the most common countries
for i, row in most_common_countries.iterrows():
    plt.text(row['Device'], row['Count'], row['Country of Origin'], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
#F
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

country_counts = df['Country of Origin'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Technological Products by Country of Origin")
plt.show()
```

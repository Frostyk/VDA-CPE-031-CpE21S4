```python
#a. Read each device's date of releases and sales. Show it using a line plot
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Date of Release'], df['Price (USD)'], alpha=0.7, color='red')
plt.title('Prices over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#b.
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('Technological-Products-Sample-Data.xlsx')
 # Convert the DataFrame to a CSV file
df.to_csv('Technological-Products-Sample-Data.csv', index=False)
df = pd.read_csv('Technological-Products-Sample-Data.csv')

plt.figure(figsize=(12, 5))
plt.plot(df['Date of Release'], df['Price (USD)'], alpha=0.7, linestyle=':', color='r', label='Price', marker='o',linewidth=3)
plt.title('Prices over Time')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.legend(loc='lower right')
plt.show()
```
